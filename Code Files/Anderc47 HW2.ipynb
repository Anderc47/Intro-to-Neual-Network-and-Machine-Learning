3.1 Fit a linear model to the data by minimizing the associated linear regression Least Squares problem using a single Newton step

```python
# extract input
x = data[:,0]
y = data[:,1]

x.shape = (len(x),1)
o = np.ones((len(x),1))
x_new = np.concatenate((o,x),axis=1)

def isolate_w(x_new,y):
    A = np.dot(x_new.T,x_new)
    b = np.dot(x_new.T,y)
    w = np.linalg.solve(A,b)
    return w

w = isolate_w(x_new,y)
lobf = w[0] + np.dot(x,w[1])
```

3.2 plot the data along with the fitted line

```python
plt.scatter(x,y)
plt.plot(x,lobf)
plt.xlabel('year')
plt.ylabel('student debt')
plt.show()
```

```python
# import the dataset. Note you need to take the log of the imported data
csvname = 'kleibers_law_data.csv'
data = np.loadtxt(csvname,delimiter=',')
x = data[:-1,:]
y = data[-1:,:] 
print('data imported')

plt.scatter(x,y)
plt.show()

x = np.log(x)
y = np.log(y)  

def model(x,w):
    a = w[0] + np.dot(x.T,w[1:])
    return a.T

def least_squares(w):
    cost = np.sum((model(x,w)-y)**2)
    cost = cost/float(y.size)
    return cost
```
